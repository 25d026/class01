```python
# prompt: 초등학교 1학년이 할 수 있는 사회 문제 맞추기 게임인데 문제가 한정적이지 않고 다 다른 문제로 나오고 게임이 끝났을 때 총 틀린 문제가 몇문제이고 정답이 몇개인지 알려줘.그리고 틀린 문제는 자세히 설명해줘.그리고 문제 정답이 비슷한 내용이면 맞게 해줘.

import random

def generate_social_studies_question():
    """
    초등학교 1학년 수준의 사회 문제와 정답을 무작위로 생성합니다.
    """
    question_types = [
        "우리 동네 장소",
        "가족 구성원",
        "우리나라 상징",
        "학교생활 규칙",
        "계절 특징"
    ]
    question_type = random.choice(question_types)

    if question_type == "우리 동네 장소":
        places = ["학교"]
        answer = random.choice(places)
        question = f"우리가 공부하러 매일 가는 곳은 어디일까요?"
        explanation = f"학교는 우리가 선생님께 배우고 친구들과 공부하는 곳이에요."
        # 비슷한 답변 허용 (예: '학습하는 곳')
        similar_answers = ["학습하는 곳"]

    elif question_type == "가족 구성원":
        members = ["엄마"]
        answer = random.choice(members)
        question = f"나를 낳아주시고 길러주신 분은 누구일까요?"
        explanation = f"엄마는 우리를 낳아주시고 사랑으로 길러주시는 분이에요."
        # 비슷한 답변 허용 (예: '어머니')
        similar_answers = ["어머니"]

    elif question_type == "우리나라 상징":
        symbols = ["태극기"]
        answer = random.choice(symbols)
        question = f"우리나라를 나타내는 국기는 무엇일까요?"
        explanation = f"태극기는 우리나라의 국기로, 우리나라를 상징해요."
        similar_answers = [] # 태극기는 고유 명사이므로 유사 답변 없음

    elif question_type == "학교생활 규칙":
        rules = ["규칙"]
        answer = random.choice(rules)
        question = f"학교에서 친구와 함께 즐겁게 지내기 위해 지켜야 할 약속은 무엇일까요?"
        explanation = f"친구와 사이좋게 지내는 것은 학교에서 즐겁게 생활하기 위한 아주 중요한게 규칙이에요."
        # 비슷한 답변 허용 (예: '친하게 지내기')
        similar_answers = ["친하게 지내기"]


    elif question_type == "계절 특징":
        seasons = ["봄", "여름", "가을", "겨울"]
        answer = random.choice(seasons)
        if answer == "봄":
            question = "꽃이 피고 날씨가 따뜻해지는 계절은 무엇일까요?"
            explanation = "봄에는 따뜻한 바람이 불고 예쁜 꽃들이 피어나요."
            similar_answers = []
        elif answer == "여름":
            question = "햇볕이 쨍쨍하고 물놀이하기 좋은 계절은 무엇일까요?"
            explanation = "여름은 날씨가 덥고 시원한 물놀이를 즐길 수 있는 계절이에요."
            similar_answers = []
        elif answer == "가을":
            question = "나뭇잎 색깔이 예쁘게 변하고 시원한 바람이 부는 계절은 무엇일까요?"
            explanation = "가을에는 나뭇잎이 빨갛고 노랗게 물들고 시원한 바람이 불어요."
            similar_answers = []
        elif answer == "겨울":
            question = "눈이 오고 날씨가 추워서 따뜻한 옷을 입는 계절은 무엇일까요?"
            explanation = "겨울은 날씨가 춥고 눈이 와서 눈싸움을 하거나 따뜻한 집에서 쉬기 좋은 계절이에요."
            similar_answers = []


    return question, answer, explanation, similar_answers

def play_game():
    """
    사회 문제 맞추기 게임을 진행합니다.
    """
    total_questions = 0
    correct_answers = 0
    incorrect_answers = []

    print("안녕하세요! 초등학교 1학년을 위한 사회 문제 맞추기 게임을 시작합니다!")

    while True:
        question, answer, explanation, similar_answers = generate_social_studies_question()
        total_questions += 1

        print("\n--------------------")
        print(f"문제 {total_questions}: {question}")
        user_answer = input("정답을 입력해주세요: ").strip()

        # 정답 확인 (대소문자 구분 없이, 비슷한 답변 허용)
        if user_answer.lower() == answer.lower() or user_answer.lower() in [sa.lower() for sa in similar_answers]:
            print("✨ 정답입니다! 정말 잘했어요! ✨")
            correct_answers += 1
        else:
            print(f"😅 아쉬워요. 정답은 '{answer}' 입니다.")
            print(f"설명: {explanation}")
            incorrect_answers.append({'question': question, 'user_answer': user_answer, 'correct_answer': answer, 'explanation': explanation})

        play_again = input("\n계속 하시겠어요? (네/아니오): ").strip().lower()
        if play_again != '네':
            break

    print("\n====================")
    print("👍 게임이 끝났습니다! 👍")
    print(f"총 문제 수: {total_questions}개")
    print(f"맞힌 정답 수: {correct_answers}개")
    print(f"틀린 문제 수: {len(incorrect_answers)}개")

    if incorrect_answers:
        print("\n👇 틀린 문제는 다음과 같습니다. 👇")
        for i, item in enumerate(incorrect_answers):
            print(f"\n--- 틀린 문제 {i+1} ---")
            print(f"문제: {item['question']}")
            print(f"내가 쓴 답: {item['user_answer']}")
            print(f"올바른 정답: {item['correct_answer']}")
            print(f"설명: {item['explanation']}")

    print("\n다음에 또 만나요! 😊")

# 게임 실행
play_game()
```
